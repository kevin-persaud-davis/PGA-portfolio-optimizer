### __Summary of Fantasy Total Points__

Fantasy total points (FTP) is the target variable of this dataset and thus an important one to understand. FTP is comprised of two main components that I will break down seperately. First, hole scoring points, is the amount of points per hole a golfer receieves over the tournament they compete in with a total of four rounds played and eighteen holes per round. Each hole is given a par score, ranging from three, four, and five. A player then tries to shoot the lowest score possible for every hole. DraftKings has a fantasy point mapping for each type of score a player can get. The table below shows the different scores a player can receive for every hole. 

|__Score Type__ | __Fantasy Points__ |
| :- | :- |
| Double eagle or better | 13 |
| Eagle | 8 |
| Birdie | 3 |
| Par | 0.5 |
| Bogey | -0.5 |
| Double or worse | -1|
